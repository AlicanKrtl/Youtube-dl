### Loss functions and optimizers: keys to configuring the learning process

* Kayıp fonksiyonu - Eğitim sırasında minimize edilen sayıdır. İşin başarı miktarını ifade eder.
* Optimizasyon fonksiyonu - Kayıp fonksiyonuna bağlı olarak ağın nasıl güncelleneceğine karar verir.

Bazı optimizasyonlar
* binary crossentropy - 2 sınıflı bir sınıflandırmada
* categorical crossentropy - 2 den fazla sınıfı bulunan sınıflandırmalarda
* mean squared error - regresyon problemlerinde 
* connectionist temporal classification( CTC ) - zaman serisi problemlerinde
