### Keras, TensorFlow, Theano, and CNTK

Bu iki model aynıdır.

```python
from keras import models 
from keras import layers

model = models.Sequential() 
model.add(layers.Dense(32,activation='relu', input_shape=(784,))) 
model.add(layers.Dense(10,activation='softmax'))
model.summary()
```

#### 7. bölümde functional API konusu daha detaylı anlatılacak. O zamana kadar Sequential() ile ilerleyeceğiz.

```python
from keras.models import Model
input_tensor = layers.Input(shape=(784,)) 
x = layers.Dense(32, activation='relu')(input_tensor) 
output_tensor = layers.Dense(10, activation='softmax')(x)
model = Model(inputs=input_tensor, outputs=output_tensor)
model.summary()
```

Hangi model seçilirse seçilsin sonrasında yapılacak işlemler farklılık göstermez.
